**Replacing Labels - Positive with 1 and Negative with 0**

```python
df.sentiment.replace("positive",1,inplace=True)
df.sentiment.replace("negative",2,inplace=True)
```

```python
df.head()
```

**Text Preprocessing**

```python
def data_processing(text):
  text = text.lower()
  text = re.sub('<br />','',text) #re.sub(pattern, repl, string, count=0, flags=0)
  text = re.sub(r"https\S+|www\S+|http\S+",'',text, flags=re.MULTILINE) #replacing the URLs
  text = re.sub(r'\@w+|\#','',text) #removing hashtags and @ symbol
  text = re.sub(r'[^\w\s]','',text) #removing punctuations
  #removing the stop words
  text_tokens = word_tokenize(text)
  filtered_text = [w for w in text_tokens if not w in stop_words]
  return " ".join(filtered_text)
```

```python
nltk.download('punkt')
```

```python
df.review = df['review'].apply(data_processing)
```

**Stemming**

```python
stemmer = PorterStemmer()
def stemming(data):
  text = [stemmer.stem(word) for word in data]
  return data
```

```python
df.review = df['review'].apply(stemming)
```

```python
df['word count'] = df['review'].apply(no_of_words)
df.head()
```

**Factorizing the Data**

```python
X = df['review']
Y = df['sentiment']
```

```python
vect = TfidfVectorizer() #Read more about TF-IDF
X = vect.fit_transform(df['review'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.3, random_state = 42)
```

```python
print("Size of x_train: ",x_train.shape)
print("Size of x_test: ",x_test.shape)
print("Size of y_train: ",y_train.shape)
print("Size of y_test: ",y_test.shape)
```

**Create an ML Model**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')
```

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
logreg_pred = logreg.predict(x_test)
logreg_acc = accuracy_score(logreg_pred,y_test)
print(logreg_acc*100)
```

```python
print(confusion_matrix(y_test,logreg_pred))
print('\n')
print(classification_report(y_test,logreg_pred))
```

```python
svc = LinearSVC()
svc.fit(x_train,y_train)
svc_pred = svc.predict(x_test)
svc_acc = accuracy_score(svc_pred,y_test)
print(svc_acc*100)
```

```python
print(confusion_matrix(y_test,svc_pred))
print('\n')
print(classification_report(y_test,svc_pred))
```
